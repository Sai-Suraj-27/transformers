```python
import torch
from transformers import AutoModel, AutoTokenizer
```

```python
print(">>> STAGE A: Generating Ground Truth...")

# 1. Load Original Model
model_id = "jinaai/jina-embeddings-v3"
original_model = AutoModel.from_pretrained(model_id, trust_remote_code=True)
original_model.eval()

# 2. Prepare Inputs
input_text = ["This is a test sentence for Jina V3."]
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
inputs = tokenizer(input_text, return_tensors="pt")

# 3. Run Forward Pass (Get Ground Truth)
with torch.no_grad():
    outputs = original_model(**inputs)
    # Jina V3 remote code returns a list/tuple, usually [last_hidden_state, pooler_output, ...]
    # We want the last_hidden_state (embeddings before pooling)
    expected_hidden_state = outputs[0]

# 4. Pack everything into a dictionary
verification_data = {
    "state_dict": {k: v.cpu() for k, v in original_model.state_dict().items()}, # Move to CPU
    "input_ids": inputs["input_ids"],
    "attention_mask": inputs["attention_mask"],
    "expected_output": expected_hidden_state
}

# 5. Save to disk
torch.save(verification_data, "jina_verification_data.pt")
print("✅ Ground truth saved to 'jina_verification_data.pt'")
```

```python
import torch
from transformers import AutoConfig
# Import your model class
from transformers.models.jina_embeddings_v3.modeling_jina_embeddings_v3 import JinaEmbeddingsV3Model

def rename_keys(key):
    """Rename keys from roberta prefix to match our model architecture."""
    if "emb_ln." in key:
        return key.replace("roberta.emb_ln.", "embeddings.LayerNorm.")
    if key.startswith("roberta.embeddings."):
        key = key.replace("roberta.embeddings.", "embeddings.")
    if "roberta.encoder.layers." in key:
        key = key.replace("roberta.encoder.layers.", "encoder.layer.")
    if "mixer.Wqkv." in key:
        key = key.replace("mixer.Wqkv.", "attention.attention_class.Wqkv.")
    if "mixer.out_proj." in key:
        key = key.replace("mixer.out_proj.", "attention.output.dense.")
    if "norm1." in key:
        key = key.replace("norm1.", "attention.output.LayerNorm.")
    if "mlp.fc1." in key:
        key = key.replace("mlp.fc1.", "intermediate.dense.")
    if "mlp.fc2." in key:
        key = key.replace("mlp.fc2.", "output.dense.")
    if "norm2." in key:
        key = key.replace("norm2.", "output.LayerNorm.")
    if "roberta.pooler." in key:
        key = key.replace("roberta.pooler.", "pooler.")
    return key

print(">>> STAGE B: Verifying Custom Implementation...")

# 1. Load Data
data = torch.load("jina_verification_data.pt")
input_ids = data["input_ids"]
attention_mask = data["attention_mask"]
expected_output = data["expected_output"]
original_weights = data["state_dict"]

# 2. Initialize Your Model
config = AutoConfig.from_pretrained("jinaai/jina-embeddings-v3", trust_remote_code=True)
my_model = JinaEmbeddingsV3Model(config)
my_model.eval()

# 3. Rename and Load Weights
new_state_dict = {}
for k, v in original_weights.items():
    new_state_dict[rename_keys(k)] = v

missing, unexpected = my_model.load_state_dict(new_state_dict, strict=True)
print(f"Load Results -> Missing: {len(missing)}, Unexpected: {len(unexpected)}")
```
